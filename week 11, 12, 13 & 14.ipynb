### 1. 2 Bar Plot

```python
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.bar(courses, values, color ='red', width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

# 5. Show an example of Skewed data and removal of skewedness

```python
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=1000)  # Generating skewed data

# Visualizing the skewed data
plt.figure(figsize=(8, 5))
plt.hist(skewed_data, bins=30, edgecolor='black')
plt.title('Skewed Data Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Applying log transformation to the skewed data
transformed_data = np.log(skewed_data + 1)  # Adding 1 to avoid log(0)

# Visualizing the transformed (less skewed) data
plt.figure(figsize=(8, 5))
plt.hist(transformed_data, bins=30, edgecolor='black', color='green')
plt.title('Transformed (Less Skewed) Data Distribution')
plt.xlabel('Transformed Values (Log)')
plt.ylabel('Frequency')
plt.show()
```

# 6. For a sales dataset do a Time Series Visualization

```python
np.random.seed(42)
dates = pd.date_range('2023-01-01', '2023-12-31')
sales = np.random.randint(50, 200, size=len(dates))

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})
sales_data.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(sales_data.index, sales_data['Sales'], label='Sales', color='blue')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

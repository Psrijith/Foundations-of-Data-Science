```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Generate right-skewed data
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=1000)

# Calculate skewness of the original data
original_skewness = skew(skewed_data)

# Visualize the original skewed data
plt.figure(figsize=(8, 4))
sns.histplot(skewed_data,  kde=True, color='skyblue')
plt.title(f'Original Skewed Data (Skewness: {original_skewness:.2f})')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Apply log transformation to reduce skewness
log_transformed_data = np.log1p(skewed_data)

# Calculate skewness of the transformed data
transformed_skewness = skew(log_transformed_data)

# Visualize the transformed data
plt.figure(figsize=(8, 4))
sns.displot(log_transformed_data,  kde=True, color='salmon')
plt.title(f'Transformed Data (Skewness: {transformed_skewness:.2f})')
plt.xlabel('Values (Log Transformed)')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic sales dataset with random dates
np.random.seed(42)
num_days = 365  # Number of days in the dataset
start_date = pd.to_datetime('2022-01-01')
date_rng = pd.date_range(start=start_date, periods=num_days, freq='D')

sales_data = pd.DataFrame({
    'Date': date_rng,
    'Sales': np.random.randint(50, 200, size=num_days)  
})

# Set the 'Date' column as the index
sales_data.set_index('Date', inplace=True)

# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(sales_data.index, sales_data['Sales'], label='Sales', color='lightblue')
plt.title('Synthetic Sales Time Series Visualization')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Sample data with two features (2D)
np.random.seed(42)
data = np.random.rand(100, 2) * 10  # Generating random 2D data

# Apply PCA to reduce dimensionality
pca = PCA(n_components=1)
reduced_data = pca.fit_transform(data)

# Original data plot
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1])
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Reduced data plot
plt.subplot(1, 2, 2)
plt.scatter(reduced_data, np.zeros_like(reduced_data))
plt.title('Reduced Dimensionality (PCA)')
plt.xlabel('Principal Component 1')

plt.tight_layout()
plt.show()
```
